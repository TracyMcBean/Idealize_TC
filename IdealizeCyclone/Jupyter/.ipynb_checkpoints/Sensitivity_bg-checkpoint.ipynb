```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#import plotly
from scipy.interpolate import griddata

import sys
sys.path.append('../../../AsymptoticSolver/')
sys.path.append('..')

from Utilities.coord_func import cart2pol
from Utilities.add_var import add_u_polar
from AsymptoticSolver import polar_dft, polar_idft, pick_fourier_comp

#%matplotlib notebook
```

```python
# Data for centerline
center_file       = "../Data/center_fiona.npy" 
center = np.load(center_file)
x_center = center[:,0]
y_center = center[:,1]

# Initial data
ds_orig = xr.open_dataset('../../../../init_data/dei4_NARVALII_2016081700_fg_DOM01_ML_0012.nc')
#ds_orig = add_u_polar(ds_orig,center)

# Number of levels in file (highest index is lowest level -> p-system)
nlev = len(ds_orig.height.values)
height = ds_orig.z_ifc.values
```

```python
da_u = ds_orig.u[0]
da_v = ds_orig.v[0]

# Horizontal wind
h_wind_da = np.sqrt(da_u**2 + da_v**2)
```

Now the regions in which to calculate the mean are set up. These regions are selected as boxes around the cyclone center. The minimal distance to the the center is chosen as 100km 

```python
r_earth  = 6371                # Earths radius
min_dist = 100 / r_earth       # Minimum distance to cyclone center in radian


lon_up   = np.linspace(ds_orig.clon.max(), x_center[-1] + min_dist, 10, endpoint=True)
lon_down = np.linspace(ds_orig.clon.min(), x_center[-1] - min_dist, 10, endpoint=True)
lat_up   = np.linspace(ds_orig.clat.min(), y_center[-1] - min_dist, 10, endpoint=True)
lat_down = np.linspace(ds_orig.clat.max(), y_center[-1] + min_dist, 10, endpoint=True)

# Extract different regions containing cyclone
lonlat_box = {'lon_up': lon_up,'lon_down':lon_down,
              'lat_up': lat_up , 'lat_down': lat_down}

bg_wind = np.empty([len(lon_up), nlev])
bg_u = np.empty([len(lon_up), nlev])
bg_v = np.empty([len(lon_up), nlev])
```

```python
lon_up[6]* 180/np.pi, lon_down[6]* 180/np.pi, lat_up[6]* 180/np.pi, lat_down[6]* 180/np.pi
lon_up[6], lon_down[6], lat_up[6], lat_down[6]
```

Calculate the background wind for every layer of every selection.

```python
var_da = h_wind_da.copy()
#var_da = da_u.copy()
#var_da = da_v.copy()
# Go over all regions and calculate the bg wind for each
for j in range(0, len(lonlat_box['lon_up'])):
    sub_da = var_da.copy()
    
    # Select region of interest
    sub_da = sub_da.where(sub_da['clon'] < lonlat_box['lon_up'][j], drop=True)
    sub_da = sub_da.where(sub_da['clon'] > lonlat_box['lon_down'][j], drop=True)
    sub_da = sub_da.where(sub_da['clat'] > lonlat_box['lat_up'][j], drop=True)
    sub_da = sub_da.where(sub_da['clat'] < lonlat_box['lat_down'][j], drop=True)
    
    for i in range(0, nlev):
        bg_wind[j,i] = sub_da.values[i].mean()
        #bg_u[j,i] = sub_da.values[i].mean()
        #bg_v[j,i] = sub_da.values[i].mean()
```

#### Bai approach

```python
# Set center as constant. Only use mean over the levels higher than 1,5km.(that's why the 14 is used)
center_loc = [center[:(len(center)-14),0].mean(), center[:(len(center)-14),1].mean()]
# Define radius smaller than 200km in radian
r_min = 200/6371
r_max = 800/6371

bg_array = np.empty((len(ds_orig.u[0]),2))

start_lev= 0
#start_lev= len(ds_orig.u[0]) -2

for lev in range(start_lev, len(ds_orig.u[0])):
    
    #select single levels
    single_lev_u = ds_orig.u[0, lev]
    single_lev_v = ds_orig.v[0, lev]
    
    dist_array = np.empty([len(single_lev_u.ncells)])

    #print('Filling in the distances...')
    #for i in range(len(single_lev_u.ncells)):
    #    dist_array[i]= np.sqrt((single_lev_u.clon.values[i] - center_loc[0])**2 \
    #              + (single_lev_u.clat.values[i] - center_loc[1])**2)

    dist_array = np.sqrt((single_lev_u.clon.values - center_loc[0])**2 \
                 + (single_lev_u.clat.values - center_loc[1])**2)
    
    #print('Selecting outer area...')
    # Create new dataset combining pres and dist data.
    circle_sel = xr.Dataset({'u': single_lev_u, 'v':single_lev_v, 'dist': (('ncells'), dist_array)})
    # Extract annular around cylcone.
    circle_sel = circle_sel.where(circle_sel['dist'] >= r_min, drop=True)
    circle_sel = circle_sel.where(circle_sel['dist'] <= r_max, drop=True)

    
    bg_array[lev] = [circle_sel.u.mean(), circle_sel.v.mean()]
```

```python
plt.tripcolor(circle_sel.clon, circle_sel.clat, circle_sel.u, shading='gouraud')
```
