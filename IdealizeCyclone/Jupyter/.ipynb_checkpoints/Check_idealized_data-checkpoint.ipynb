```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import plotly
```

```python
%matplotlib notebook
```

```python
# Path of nc file to be read
file_path = '../../../../init_data/'
file_name = 'temp_thetav_idealized.nc'
# Read in data
ds = xr.open_dataset(file_path+file_name)
```

```python
# Create plot of level (suitable for temperature plotting with the coolwarm colormap)
ax = plt.subplot(111)
plt.tripcolor(ds.clon, ds.clat, ds.temp.isel( time=0, height=-1), cmap='coolwarm')
plt.title('Temperature in 20m')
plt.xlabel('Longitude rad')  
plt.ylabel('Latitude rad')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

cbar = plt.colorbar() 
cbar.ax.set_ylabel('Temp in K')
```
