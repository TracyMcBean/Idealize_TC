```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import plotly
```

```python
%matplotlib notebook
```

```python
# Path of nc file to be read
file_path = '../../../../init_data/'
file_name = 'init_idealized_DOM01_2016081712.nc'
# Read in data
#ds_ideal = xr.open_dataset(file_path+file_name)
ds_ideal = xr.open_dataset('/scratch/usr/bekthkis/ICON_08_2019/Fiona2016/init_data/init_idealized_DOM01_2016081712.nc')
```

```python
# Create plot of level (suitable for temperature plotting with the coolwarm colormap)
ax = plt.subplot(111)
plt.tripcolor(ds_ideal.clon, ds_ideal.clat, ds_ideal.temp.isel( time=0, height=60), cmap='coolwarm')
plt.title('Temperature in %s' % ds_ideal.z_ifc[60])
plt.xlabel('Longitude rad')  
plt.ylabel('Latitude rad')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

cbar = plt.colorbar() 
cbar.ax.set_ylabel('Temp in K')
```
